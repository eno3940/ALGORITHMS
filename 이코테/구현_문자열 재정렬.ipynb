```python
from collections import deque
import copy
```

```python
Strings=list(input())
```

```python
Strings.sort()
```

```python
Strings
```

```python
Strings=deque(Strings)
```

```python
number=0
for string in copy.deepcopy(Strings):
    if string.isdigit():
        number+=int(string)
        Strings.popleft()
```

```python
number
```

```python
Strings.append(str(number))
```

```python
Strings=''.join(Strings)
```

```python
def solve(P, I):
    answer = 0
    cnt=0
    from collections import deque
    P=deque(P)
    for i in range(len(I)):
        try:
            while P[i]!=I[i]:
                del P[i]
                cnt+=1
        except:
            pass
    if ''.join(list(P))==I:
        answer=cnt
    if cnt==0 and len(P)-len(I)>0:
        answer=len(P)-len(I)
    else:
        answer="IMPOSSIBLE"
    return answer

num_cases = int(input())

for case in range(1, num_cases + 1):
    I = input()
    P = input()
    y = solve(P, I)
    print('Case #{}: {}'.format(case, y))
```

```python
def solve(P, I):
    from collections import deque
    import copy
    answer = 0
    new_list=deque()
    del_list=deque()
    P=deque(P)
    for i in I:
        for p in copy.deepcopy(P):
            if i==p:
                new_list.append(P.popleft())
                break
            else:
                del_list.append(P.popleft())
    if ''.join(new_list)==I:
        answer=len(P)+len(del_list)
    else:
        answer="IMPOSSIBLE"
    return answer

num_cases = int(input())

for case in range(1, num_cases + 1):
    I = input()
    P = input()
    y = solve(P, I)
    print('Case #{}: {}'.format(case, y))
```
